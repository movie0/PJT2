```python
df_new=df_move.transpose()[2:].reset_index()
df_new=df_new.rename(columns=df_new.iloc[0])
df_new=df_new.drop(df_new.index[[0]])
```

```python
origin=pd.DataFrame()
result=pd.DataFrame()

for i in range(int(len(df_new.columns)/8)):
    for j in range(len(df_new)):
        start=0+(8*i)
        end=8+(8*i)
        temp=df_new.iloc[j,start:end].to_frame().transpose()
        result=pd.concat([origin,temp])
        origin=result
    origin=result
```
